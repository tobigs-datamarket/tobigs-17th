### 1-5) 마지막으로 고유 벡터 축으로 값을 변환합니다


https://docs.scipy.org/doc/numpy/reference/generated/numpy.concatenate.html

```python
def new_coordinates(X,eigenvectors):
    for i in range(eigenvectors.shape[0]):
        if i == 0:
            new = [X.dot(eigenvectors.T[i])]
        else:
            new = np.concatenate((new, [X.dot(eigenvectors.T[i])]), axis = 0)
    return new.T

# 모든 고유 벡터 축으로 데이터를 projection한 값입니다
```

```python
from sklearn.preprocessing import StandardScaler

def MYPCA(X,number):
    # 스케일링
    scaler = StandardScaler()
    X_std = scaler.fit_transform(X)
    # 공분산 행렬 계산
    cov_matrix = np.cov(X_std.T)
    
    # 고유값과 고유벡터 계산
    eigenvalues = lin.eig(cov_matrix)[0]
    eigenvectors = lin.eig(cov_matrix)[1]
    
    # 고유값 정렬 및 고유값 크기에 따라 고유벡터도 정렬 ??
    # 원래 정렬해서 답이 나오는데 꼭해야한는 건가요?
    def new_coordinates_2(X,eigenvectors,number):
        for i in range(number):
            if i == 0:
                new = [X.dot(eigenvectors.T[i])]
            else:
                new = np.concatenate((new, [X.dot(eigenvectors.T[i])]), axis = 0)
        return new.T
    
    
    # number만큼 주성분 축의 개수 설정
    
    
    return new_coordinates_2(X_std,eigenvectors,number) #새로운 축으로 변환되어 나타난 데이터 return
```

## 4) PCA 외 수업에서 다룬 차원축소 기법(FS,GA,LDA,SNE 등) 중 택1하여 밑바닥부터 만들어보기

```python
## 전진 선택법

def forward(x,y, model):
    feature_list = x.columns.tolist()
    
    best = (0,0)
    binn = []
    
    for _ in range(len(feature_list)): #최대 포함가능한 변수만큼 반복
        
        for num, feature in enumerate(feature_list): #각 변수를 합해 점수계산 가장높은걸 더함
            model.fit(x.loc[binn+[feature]],y)
            if best[1] < model.score(x) :
                best = (feature,model.score(x))
        
        if best[1] == 0 : #만약 성능 향상에 도움을 못준 경우 종료
            return binn
                
        binn.append(feature) #포함할 핏쳐는 binn에 저장
        feature_list.remove(feature)  
        best = (0,0) #초기화
        
    return binn
            
    
```
